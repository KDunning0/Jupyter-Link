```python
#Import mumpy and pyplot
import numpy
import matplotlib.pyplot as mplot
```

```python
#Use numpy to generate 500 chi-square random numbers with 5 degrees of freedom.
chisq = numpy.random.chisquare(df=5, size=500)
```

```python
#Create and show histogram, get/create current axes, add bins with colors.
(mplot.figure(figsize=(8, 5))
 .gca()
 .hist(chisq, bins=30, density=True, alpha=0.7, color="purple"))
mplot.show()
```
