```python
#Import numpy and matplotlib
import numpy
import matplotlib.pyplot as mplot
```

```python
# Creating 500 random numbers of poisson distribution with 25 events/lamda
poissonval = numpy.random.poisson(lam=25, size=500)
```

```python
# Create and show histogram, get/create current axes, add and arrange bins with colors.
(mplot.figure(figsize=(8, 5))
 .gca()
 .hist(poissonval, bins=numpy.arange(0, 50, 0.75), density=True, alpha=0.7, color="orange"))
mplot.show()
```
