```python
import pandas
import numpy
import matplotlib.pyplot as mplot
```

```python
#Read the file with pandas into a dataframe.
Studentdf = pandas.read_csv('2023-06-13-survey.csv')
```

```python
#Create the first histogram, Operating System
Studentdf['Operating System'].hist()

# Add title and labels
mplot.title('Student Operating Systems')
mplot.xlabel('Operating System')
mplot.ylabel('Count')

# Display the plot
mplot.show()
```

```python
# Create the second histogram, CPU Number of Cores (int).
Studentdf['CPU Number of Cores (int)'].hist(edgecolor='black')

# Add title and labels
mplot.title('Student CPU Number of Cores (int)')
mplot.xlabel('CPU Number of Cores (int)')
mplot.ylabel('Count')

# Display the plot
mplot.show()
```

```python
# Create the third histogram, CPU Cycle Rate in GHz.
Studentdf['CPU Cycle Rate (in GHz)'].hist(bins=numpy.arange(0, 5, 0.1), edgecolor='black')

# Add title and labels
mplot.title('Student CPU Cycle Rate (in GHz)')
mplot.xlabel('CPU Cycle Rate (in GHz)')
mplot.ylabel('Count')
mplot.xlim(0, 5)

# Display the plot
mplot.show()
```

```python
# Create the fourth histogram, Hard Drive Size (in GB).
Studentdf['RAM (in GB)'].hist(bins=numpy.arange(0, 66, 1), edgecolor='black')

# Add title and labels
mplot.title('Student RAM (in GB)')
mplot.xlabel('RAM (in GB)')
mplot.ylabel('Count')

# Display the plot
mplot.show()
```

```python
# Create the fifth Histogram: Hard Drive
Studentdf['Hard Drive Size (in GB)'].hist(bins=numpy.arange(0, 4100, 50), edgecolor='black')

# Add title and labels
mplot.title('Student Hard Drive Size (in GB)')
mplot.xlabel('Hard Drive Size (in GB)')
mplot.ylabel('Count')
mplot.xticks(numpy.arange(0, 4100, 400))

# Display the plot
mplot.show()
```

```python
# Create the Sixth Histogram: CUDA Cores
Studentdf['GPU CUDA Number of Cores (int)'].hist(bins=numpy.arange(0, 6000, 100), edgecolor='black')

# Add title and labels
mplot.title('Student GPU CUDA Number of Cores (int)')
mplot.xlabel('GPU CUDA Number of Cores (int)')
mplot.ylabel('Count')

# Display the plot
mplot.show()
```
