```python
#Import necessary libraries
import numpy
import pandas
import seaborn as sb
import matplotlib.pyplot as mplot
```

```python
#Set up the sample total and seed for all runs.
samplenum = 1000
numpy.random.seed(42)
```

```python
#First value of n is 20.  Histogram number 1.
binomdat1 = numpy.random.binomial(n=20, p=0.5, size=samplenum)
normdat1 = numpy.random.normal(loc=10, scale=2, size=samplenum)
dataframe1 = pandas.DataFrame({
    "Numbers": numpy.concatenate([binomdat1, normdat1]),
    "Source": numpy.repeat(["Binomial", "Normal"], samplenum)
})
sb.histplot(data=dataframe1, x="Numbers", hue="Source", alpha=0.5, bins=30)
mplot.title("N=20 Binomial and Normal Distribution data")
```

```python
#Second value of n is 100.  Histogram number 2.
binomdat2 = numpy.random.binomial(n=100, p=0.5, size=samplenum)
normdat2 = numpy.random.normal(loc=50, scale=5, size=samplenum)
dataframe2 = pandas.DataFrame({
    "Numbers": numpy.concatenate([binomdat2, normdat2]),
    "Source": numpy.repeat(["Binomial", "Normal"], samplenum)
})
sb.histplot(data=dataframe2, x="Numbers", hue="Source", alpha=0.5, bins=30)
mplot.title("N=100 Binomial and Normal Distribution data")
```

```python
#Third value of n is 1000.  Histogram number 3.
binomdat3 = numpy.random.binomial(n=1000, p=0.5, size=samplenum)
normdat3 = numpy.random.normal(loc=500, scale=15, size=samplenum)
dataframe3 = pandas.DataFrame({
    "Numbers": numpy.concatenate([binomdat3, normdat3]),
    "Source": numpy.repeat(["Binomial", "Normal"], samplenum)
})
sb.histplot(data=dataframe3, x="Numbers", hue="Source", alpha=0.5, bins=30)
mplot.title("N=1000 Binomial and Normal Distribution data")
```
